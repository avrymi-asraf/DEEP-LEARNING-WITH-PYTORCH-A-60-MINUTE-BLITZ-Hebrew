# מטא למידה
כאשר אנחנו בונים רשת נוירונים, עיקר הלמידה של הרשת הוא דרך האלגוריתם של גרדיאנט דסנט, אבל אם נחשוב, יש הרבה מטא למידה שקורית. כל הייפר פרמטרים, מבנה הרשת- אלו מאפיינים של הרשת שלנו ששהלמידהה שרואית בהם ההיא ע"י בני אדם. 
אנחנו מנסים מספר מנה רשתות, ובוחרים את המבנה הרצוי, או את הייפר פרמטרים הרצויים. 

ללמידה כזאת נקרא מטא למידה או למידה אקראית כיון שהיא לא קורית ע"י האלגוריתם של גרדיאנט דסנט.
אפשר לעשות כמה וכמה אלגוריתמים של למידה אקראית שבמקום שיפעלו ע"י בני אדם יפעלו ע"י מכונה.

## כיוונים

- ברירת מודלים, אפשר להריץ כמה מודלים, ורק מי שמתחיל טוב להמשיך איתו. אם אנחנו רואים שהרשת תלויה בנתונים האקריים של ההתחלה, אז ברירת מודלים יכולה להיות פתרון טוב.
- מנגנון גנטי, לנתונים קבועים, כמו מבנה הרשת, או אופטיממיזר או פונקציית הערכה אפשר לתכנן אלגוריתם שמריץ מספר מנגנונים בו זמנית ובורר את הטובים מהניהם, משכפל אותם וממשיך שוב. שאלה שעולה במקרב כזה - האם קיימת דרך להוסי על מודל קיים? או שכל מודל צריך להתחיל מהתחלה?, אם כל מודל צריך  להתחיל מהתחלה, זה דורש משימת אימון קצרה. 
- חזרה בכיון, אם נשמור את הפרמטרים של מספר הצעדים האחרונים, עם הערכים שלהם, נוכל לדעת האם התקדמות בכיון מסויים הועילה או לא. אם היא לא הועילה ננסה למצוא כיון אחר. מה שכ ןצריך למצוא כיון שונה ממש להתקדמות ולא שונה רק קצת, השאלה איזו מטריקה תאפשר לנו להעריך את ההבדל בן גרדיאנט של שלב זה, לגרדיאנט אחר.
- הערכת ביצועים, שאלה מרכזית היא איך ניתן להעריך את הביצועים של מודל מסויים, במיוחד אם אנחנו רוצים להעדיף בן רשתות שנמצאות רק בתחילת האימון, שיטה אחת היא לקבוע רף מסויים של שמי שלא עובר אותו לא מתקדם, לחילופין, בברירת מודלים או במודל גנטי, ניתן לקבוע קבוצה, ולקחת את הקבוצה הטובה ביותר, ניתן להכניס בניקוד של כל מודל את השונות שלו מהמודלים הארחים, אם אפשר למודד זאת.
